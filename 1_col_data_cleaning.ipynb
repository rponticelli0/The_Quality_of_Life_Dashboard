```python
import pandas as pd
```

# Read Cost of Living CSV File and Begin Cleaning Data

```python
# Read csv file
col_data_df = pd.read_csv('Resources/Cost_of_Living_Docs/cost_of_living.csv')
col_data_df.head()
```

```python
# Create a copy of the original DataFrame
col_data_df_copy = col_data_df.copy()
```

```python
# Drop the 'county_fips', 'case_id', 'housing', 'food_monthly', 'transportation_monthly', 'healthcare_monthly', 'misc_monthly', 'childcare_monthly', 'taxes_monthly', 'total_monthly' columns from copy
col_data_df_copy = col_data_df_copy.drop(columns=['county_fips', 'case_id', 'housing', 'food_monthly', 'transportation_monthly', 'healthcare_monthly', 'misc_monthly', 'childcare_monthly', 'taxes_monthly', 'total_monthly'])

# Verify the column has been dropped
col_data_df_copy.head()
```

```python
# Save the updated DataFrame to a new CSV file
col_data_df_copy.to_csv('Resources/Cost_of_Living_Docs/updated_col_data.csv', index=False)
```
