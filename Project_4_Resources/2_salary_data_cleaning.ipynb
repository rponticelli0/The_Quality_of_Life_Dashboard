```python
import pandas as pd
```

# Read Average Salary per Occupation CSV File and Begin Cleaning Data

```python
# Read csv file
salary_data_df = pd.read_csv('Resources/OEWS_Docs/salary_data.csv')
salary_data_df.head()
```

```python
# Create a copy of the original DataFrame
salary_data_df_copy = salary_data_df.copy()
```

```python
salary_data_df_copy.columns
```

```python
# Drop the  columns from copy
salary_data_df_copy = salary_data_df_copy.drop(columns=['area', 'area_type', 'naics', 'naics_title', 'i_group', 'own_code', 'o_group'])

# Verify the column has been dropped
salary_data_df_copy.head()
```

```python
# Apply the string split method to remove everything after the comma
salary_data_df_copy['county'] = salary_data_df_copy['county'].str.split(',').str[0]

salary_data_df_copy.head()
```

```python
# Save the updated DataFrame to a new CSV file
salary_data_df_copy.to_csv('updated_salary_data.csv', index=False)
```
