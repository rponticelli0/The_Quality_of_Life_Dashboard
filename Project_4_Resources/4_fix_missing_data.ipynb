# Fix Identified Problem of Inconsistent "county" Data:

```python
# Standardize county naming
col_df_copy['county'] = col_df_copy['county'].str.lower().str.replace('county', '').str.strip()
salary_df_copy['county'] = salary_df_copy['county'].str.lower().str.replace('county', '').str.strip()

# Ensure state_abv is upper case for consistency
col_df_copy['state_abv'] = col_df_copy['state_abv'].str.upper()
salary_df_copy['state_abv'] = salary_df_copy['state_abv'].str.upper()
```

### This next code block can take 5+ minutes to run

```python
def find_best_match_rapid(row):
    # Limit comparison to counties within the same state
    state_counties = salary_df_copy[salary_df_copy['state_abv'] == row['state_abv']]['county']
    # Apply fuzzy matching within the state's county list
    return process.extractOne(row['county'], state_counties)[0] if not state_counties.empty else None

# Apply the fuzzy matching function
col_df_copy['best_match'] = col_df_copy.apply(find_best_match_rapid, axis=1)
```

# Merge DataFrames and Upload New CSV File:

```python
# Perform the merge using the best_match as the key
final_df = pd.merge(col_df_copy, salary_df_copy, left_on=['best_match', 'state_abv'], right_on=['county', 'state_abv'], how='inner')
```

```python
# Identify unmatched data in 'county'
unmatched_counties = col_df_copy[~col_df_copy['county'].isin(final_df['best_match'])]
print(unmatched_counties)
```
